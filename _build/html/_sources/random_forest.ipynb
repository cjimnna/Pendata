## **Implementasi Algoritma Random Forest pada Dataset Iris**

---

### 1. Pendahuluan dan Landasan Teori
Analisis ini berfokus pada klasifikasi spesies bunga Iris menggunakan algoritma Random Forest. Dalam dunia *Machine Learning*, Random Forest adalah algoritma *Ensemble Learning* yang bekerja dengan membangun banyak pohon keputusan (*Decision Trees*) secara independen selama masa pelatihan.

Mengapa menggunakan Random Forest?
* Akurasi Tinggi: Dengan menggabungkan prediksi dari banyak pohon (Bootstrap Aggregating), model ini lebih stabil.
* Anti-Overfitting: Dibandingkan satu pohon keputusan tunggal, Random Forest lebih tahan terhadap *noise* dalam data.
* Feature Importance: Membantu mengidentifikasi variabel mana yang paling berpengaruh terhadap klasifikasi.

---

### 2. Deskripsi Karakteristik Data (Data Understanding)
Tahap ini krusial untuk memastikan kualitas input sebelum masuk ke model.
* Fitur Prediktor: Panjang dan lebar kelopak (Sepal) serta mahkota (Petal).
* Variabel Target: Spesies Iris (Setosa, Versicolor, Virginica).
* Kualitas Data: Data bersifat bersih (0 missing values) dengan total 150 baris sampel.

### 4. Bedah Hasil dan Evaluasi Model

#### A. Logika Struktur Decision Tree
Untuk membedah "Black Box" dari Random Forest, kita dapat melihat struktur salah satu pohon yang terbentuk.

>forest decision tree view.png
> Analisis Teknis: Perhatikan bagaimana pohon melakukan *Splitting*. Biasanya, fitur `Petal Width` atau `Petal Length` muncul di puncak (Root Node). Hal ini secara teknis menunjukkan bahwa fitur tersebut memiliki Information Gain atau Gini Impurity terkecil, menjadikannya pembeda paling kuat antar spesies.

#### B. Evaluasi Akurasi via Confusion Matrix
Keberhasilan model diukur melalui perbandingan prediksi vs realitas.

> forest scorer 1.png

> Analisis Fungsional:
> * Diagonal Utama: Menunjukkan jumlah prediksi yang benar (True Positives).
> * Akurasi: Berdasarkan gambar, model mencapai tingkat akurasi yang hampir sempurna. Ini membuktikan bahwa kombinasi fitur morfologi bunga Iris sangat ideal untuk diklasifikasikan dengan metode Random Forest.

#### C. Kurva ROC (Receiver Operating Characteristic)
Kurva ini menggambarkan kemampuan diskriminasi model.

> forest roc.png

> Konsep: Semakin kurva mendekati pojok kiri atas (AUC mendekati 1.0), maka model semakin sempurna dalam membedakan antar kelas.
> Fungsi: Memberikan bukti visual bahwa model memiliki tingkat kesalahan *false positive* yang sangat rendah pada ketiga spesies Iris.
