## **Analisis Regresi Linier**

### Pendahulu
Regresi linier merupakan metode statistik yang digunakan untuk mengetahui hubungan antara variabel independen (x) dan variabel dependen (y).  
Metode ini digunakan untuk mencari garis terbaik yang dapat merepresentasikan pola dari data yang diberikan.

Pada proyek ini dilakukan analisis regresi linier menggunakan:

1. GeoGebra
2. Python dengan library `sklearn`
3. Perhitungan analitik sederhana

### Dataset

Data yang digunakan pada percobaan ini adalah sebagai berikut:

| Titik | x | y |
|---|---|---|
| A | 2 | 2 |
| B | 4 | 3 |
| C | 5 | 5 |
| D | 3 | 4 |
| E | 3 | 3 |
| F | 4 | 5 |
| G | 5 | 6 |


### Hasil Regresi dari GeoGebra
Screenshot_20.png


Dari hasil perhitungan GeoGebra diperoleh persamaan garis regresi:

\[
y = 1.08x
\]


Dengan:

- Koefisien regresi / slope (\(b\)) = 1.08
- Intercept (\(a\)) = 0

Model regresi ini menggunakan garis yang melewati titik asal \((0,0)\).


### Implementasi Menggunakan Python dan Sklearn



```python
# Import library
from sklearn.linear_model import LinearRegression
import numpy as np

# Data x dan y
x = np.array([2, 4, 5, 3, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 4, 3, 5, 6])

# Membuat model regresi linier tanpa intercept
model = LinearRegression(fit_intercept=False)

# Melatih model menggunakan data
model.fit(x, y)

# Mengambil nilai koefisien regresi
b = model.coef_[0]

# Menampilkan hasil
print("Koefisien regresi =", b)

# Menampilkan persamaan regresi
print(f"Persamaan regresi: y = {b:.2f}x")
```

```python
# Import library
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data x dan y
x = np.array([2, 4, 5, 3, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 4, 3, 5, 6])

# Membuat model regresi tanpa intercept
model = LinearRegression(fit_intercept=False)

# Training model
model.fit(x, y)

# Prediksi garis regresi
y_pred = model.predict(x)

# Koefisien regresi
b = model.coef_[0]

# Menampilkan hasil
print("Koefisien regresi =", b)
print(f"Persamaan regresi: y = {b:.2f}x")

# Membuat grafik
plt.figure(figsize=(8,5))

# Scatter plot data
plt.scatter(x, y, label='Data Asli')

# Garis regresi
plt.plot(x, y_pred, label='Garis Regresi')

# Judul dan label
plt.title('Regresi Linier')
plt.xlabel('X')
plt.ylabel('Y')

# Menampilkan legenda
plt.legend()

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```
