# Pengantar Data Mining
Data mining adalah bidang ilmu yang mempelajari cara menemukan pengetahuan atau informasi berharga dari kumpulan data yang besar.

Secara sederhana, data mining dapat dianalogikan seperti proses menambang emas. Dalam penambangan emas, seseorang harus memisahkan emas dari tanah dan batuan yang tidak bernilai. Begitu juga dalam data mining, kita berusaha menemukan informasi berharga dari data mentah yang jumlahnya sangat banyak.

Dalam konteks bisnis dan pengambilan keputusan, data mining bertujuan untuk menghasilkan informasi yang dapat dijadikan dasar dalam menentukan strategi atau tindakan tertentu.

Proses data mining melibatkan berbagai teknik, seperti:
* Statistik
* Matematika
* Kecerdasan Buatan (Artificial Intelligence)
* Machine Learning

Teknik-teknik tersebut digunakan untuk mengekstraksi dan mengidentifikasi pola atau pengetahuan yang bermanfaat dari database berukuran besar.

##### Keunggulan Data Mining

Dalam statistik tradisional, data yang sudah ada di masa lalu digunakan untuk memperkirakan kejadian di masa depan. Cara ini biasanya memakai rumus atau model matematika yang sudah ditentukan sebelumnya.

Sedangkan dalam data mining, analisis tidak hanya melihat data lama, tetapi juga mencoba menemukan pola tersembunyi yang sebelumnya tidak terlihat. Data mining bisa membuat model yang tidak hanya menjelaskan apa yang sudah terjadi, tetapi juga memperkirakan apa yang kemungkinan akan terjadi di masa depan.

Secara sederhana, keunggulan data mining adalah:

* Membantu memahami data lebih dalam karena dapat menemukan pola yang tidak terlihat sebelumnya.
* Membantu membuat prediksi untuk mendukung pengambilan keputusan.
* Menjadi dasar dalam menyusun strategi atau rencana ke depan.
* Membuat keputusan menjadi lebih baik karena berdasarkan hasil analisis data.

## 1. Konsep Dasar Data Mining

Data mining adalah proses menggali informasi bernilai dari kumpulan data dalam jumlah besar. Tujuannya bukan sekadar melihat data, tetapi menemukan pola, hubungan, atau pengetahuan tersembunyi yang bisa dimanfaatkan.

Secara sederhana, data mining dapat diibaratkan seperti proses menambang. Dari sekumpulan material mentah, kita berusaha menemukan bagian yang paling berharga.

Dalam konteks bisnis maupun penelitian, data mining digunakan untuk:
- Mendukung pengambilan keputusan
- Membantu membuat prediksi
- Menyusun strategi berbasis data

Proses ini memanfaatkan berbagai pendekatan, seperti:
- Statistik
- Matematika
- Artificial Intelligence
- Machine Learning

---

## 2. Perbedaan Statistik Tradisional dan Data Mining

Statistik tradisional biasanya menggunakan model atau rumus yang sudah ditentukan untuk menguji hipotesis.

Sedangkan data mining lebih bersifat eksploratif, yaitu mencari pola langsung dari data tanpa selalu bergantung pada asumsi awal. Pendekatan ini memungkinkan ditemukannya pola yang sebelumnya tidak terlihat.

Keunggulan data mining:
- Menemukan pola tersembunyi
- Membantu melakukan prediksi
- Mendukung pengambilan keputusan berbasis data
- Membantu perencanaan jangka panjang

# Memahami Data

## 1. Pengertian Data

Data adalah sekumpulan fakta mentah yang dapat diolah menjadi informasi bermakna.

Sumber data bisa berasal dari:
- Manusia
- Mesin atau sensor
- Media sosial
- Sistem transaksi
- Perangkat IoT

Memahami jenis dan karakteristik data sangat penting sebelum melakukan analisis.

---

## 2. Jenis-Jenis Data

### A. Data Terstruktur

Data terstruktur adalah data yang tersusun dalam bentuk tabel (baris dan kolom), seperti pada Excel atau database.

Contoh:
- Data mahasiswa
- Data pelanggan
- Data transaksi penjualan

Dalam tabel, setiap kolom disebut atribut atau variabel.

#### Jenis Atribut:

**1. Nominal**  
Kategori tanpa urutan.  
Contoh: jurusan, warna, jenis kelamin.

**2. Biner**  
Memiliki dua nilai saja.  
Contoh: 0/1, Ya/Tidak.

**3. Ordinal**  
Memiliki urutan, tetapi jaraknya tidak pasti.  
Contoh: rendah – sedang – tinggi.

**4. Numerik**  
Berupa angka yang dapat dihitung.  
Contoh: umur, tinggi badan, nilai ujian.

---

### B. Data Tidak Terstruktur

Data yang tidak memiliki format tabel tetap.

Contoh:
- Email
- Artikel
- Dokumen teks
- Chat

Biasanya perlu diproses terlebih dahulu sebelum dianalisis.

---

### C. Data Bahasa Alami

Data dalam bentuk teks bahasa manusia.

Contoh:
- Komentar media sosial
- Review produk
- Berita

Pengolahannya menggunakan teknik Natural Language Processing (NLP).

---

### D. Data Buatan Mesin

Data yang dihasilkan secara otomatis oleh sistem atau perangkat.

Contoh:
- Log server
- Data sensor
- Data IoT

---

### E. Data Berbasis Graph

Data yang menunjukkan hubungan antar objek.

Contoh:
- Jaringan pertemanan
- Relasi antar pengguna

---

### F. Data Multimedia

Berupa:
- Gambar
- Audio
- Video

Memerlukan teknik khusus untuk diproses.

---

### G. Data Streaming

Data yang mengalir terus-menerus secara real-time.

Contoh:
- Transaksi online
- Data sensor per detik

# DISTRIBUSI DATA

Distribusi data menunjukkan bagaimana nilai dalam dataset tersebar.

Salah satu distribusi yang sering ditemui adalah distribusi normal (berbentuk kurva lonceng), di mana sebagian besar nilai berada di sekitar rata-rata.

Rumus distribusi normal:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}}
e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Keterangan:
- $\mu$ = rata-rata
- $\sigma$ = standar deviasi
- $\sigma^2$ = variansi
- $\pi$ = 3.14159
- $e$ = 2.718

Distribusi membantu kita memahami:
- Keseimbangan data
- Kemiringan distribusi
- Keberadaan nilai ekstrem (outlier)

---

# Statistik Deskriptif

Statistik deskriptif digunakan untuk merangkum data numerik.

## A. Ukuran Pemusatan

### 1. Mean

$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

Keterangan:
- $\bar{x}$ = rata-rata
- $n$ = jumlah data
- $x_i$ = data ke-i

---

### 2. Median

Jika jumlah data ganjil:

$$
\text{Median} = x_{\frac{n+1}{2}}
$$

Jika jumlah data genap:

$$
\text{Median} = \frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2}
$$

Median data berkelompok:

$$
Me = L + \left( \frac{\frac{n}{2} - F}{f} \right) p
$$

---

### 3. Modus

Nilai yang paling sering muncul dalam data.

---

## B. Ukuran Penyebaran

### 1. Range

$$
\text{Range} = X_{max} - X_{min}
$$

---

### 2. Kuartil

- $Q_1$ = kuartil bawah  
- $Q_2$ = median  
- $Q_3$ = kuartil atas  

---

### 3. Interquartile Range

$$
IQR = Q_3 - Q_1
$$

---

### 4. Variansi

Populasi:

$$
\sigma^2 = \frac{1}{N} \sum (x_i - \mu)^2
$$

Sampel:

$$
s^2 = \frac{1}{n-1} \sum (x_i - \bar{x})^2
$$

---

### 5. Standar Deviasi

Populasi:

$$
\sigma = \sqrt{\sigma^2}
$$

Sampel:

$$
s = \sqrt{s^2}
$$

---

# Skewness (Kemencengan)

Skewness menunjukkan arah kemiringan distribusi data.

## 1. Skewness Positif
- Ekor lebih panjang di kanan
- Mean > Median > Modus

## 2. Skewness Negatif
- Ekor lebih panjang di kiri
- Mean < Median < Modus

## 3. Distribusi Simetris
- Mean = Median = Modus

Rumus skewness populasi:

$$
\gamma_1 = \frac{\frac{1}{N} \sum (x_i - \mu)^3}{\sigma^3}
$$

Interpretasi:
- $\gamma_1 > 0$ → miring ke kanan
- $\gamma_1 < 0$ → miring ke kiri
- $\gamma_1 = 0$ → simetris
