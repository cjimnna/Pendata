## **PreProcessing**

### 1. Pengertian PreProcessing
Preprocessing data adalah proses awal dalam data mining atau machine learning yang bertujuan untuk menyiapkan data mentah agar dapat digunakan dalam proses analisis atau pemodelan.

Data yang diperoleh dari berbagai sumber biasanya masih memiliki beberapa masalah seperti:
* Data tidak lengkap (missing value)
* Data tidak konsisten
* Data memiliki skala yang berbeda
* Data mengandung noise

Oleh karena itu diperlukan preprocessing agar data menjadi lebih bersih, konsisten, dan mudah dianalisis.

### 2. Tahapan Preprocessing Data

Beberapa tahapan preprocessing yang umum dilakukan yaitu:


1. Data Clening

    Menghapus data yang tidak lengkap atau memperbaiki data yang salah.

2. Data Integration

    Menggabungkan data dari berbagai sumber.

3. Data Transformation

    Mengubah data ke bentuk yang sesuai untuk analisis.

4. Data Normalization

    Menyamakan skala nilai data agar tidak ada atribut yang mendominasi.


### 3. Normalisasi Data

Normalisasi data adalah proses mengubah nilai data ke dalam skala tertentu tanpa mengubah hubungan antar data.

Tujuan normalisasi:
* Menyamankan skala data
* Menghindari dominasi nilai yang benar
* Mempermudah proses perhitungan algoritma seperti KNN, K-Mens, dan Machine Learning

Beberapa metode normalisa yang sering digunakan adalah:
* Min-Max Normalization
* Z-Score Normalization
* Decimal Scalling Normalization

---------------------------------------------

**1. Min-Max Normalization**

Min-Max Normalization mengubah nilai data ke dalam rentang 0 sampai 1.

$$
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
$$

Keterangan

- $X$ = nilai data asli  
- $X_{min}$ = nilai minimum data  
- $X_{max}$ = nilai maksimum data  
- $X'$ = nilai hasil normalisasi

Contoh Perhitungan

Data

| No |   PO    | IPK |
|----|---------|-----|
| 1  | 2000000 |  2  |
| 2  | 3000000 |  3  |
| 3  | 3000000 |  4  |
| 4  | 2000000 |  3  |

Misalkan data IPK:

2,3,4

min = 2

max = 4

Jika ingin menormalisasi nilai 3:

$$(3-2)/(4-2) = 1/2 = 0.5$$

Hasil normalisasi:

| Nilai Asli | Normalisasi |
|------------|-------------|
|        2   | 0         |
| 3     | 0.5         |
| 4     | 1         |


**2. Z-Score Normalization**

Z-Score Normalization digunakan untuk mengubah data berdasarkan rata-rata (mean) dan standar deviasi.

$$
Z = \frac{X - \mu}{\sigma}
$$

**Keterangan**

- $X$ = nilai data  
- $\mu$ = rata-rata data  
- $\sigma$ = standar deviasi  
- $Z$ = nilai normalisasi

Contoh Perhitungan

Data = 2,3,4

Mean:

$$(2+3+4)/3=3$$

Standar deviasi

$$\sigma = 0.816$$

Normalisasi untuk nilai 3:

$$(3-3)/0.816 = 0$$

**3. Decimal Scaling Normalization**

Metode ini menormalisasi data dengan cara membagi nilai data dengan pangkat 10.

$$
X' = \frac{X}{10^j}
$$

**Keterangan**

- $X$ = nilai data  
- $j$ = jumlah digit terbesar  
- $X'$ = nilai normalisasi

Contoh Perhitungan

Data:

200000,300000,400000

Nilai terbesar = 400000

Jumlah digit = 6

$$j = 6$$

Normalisasi:

200000 / $10^6$ = 0.2

300000 / $10^6$ = 0.3

400000 / $10^6$ = 0.4

Hasil:

| Nilai Asli | Normalisasi |
|------------|-------------|
| 200000     | 0.2         |
| 300000     | 0.3         |
| 400000     | 0.4         |

### 4. Kesimpulan

Preprocessing merupakan tahap penting dalam pengolahan data sebelum dilakukan analisis. Salah satu proses penting dalam preprocessing adalah **normalisasi data** untuk menyamakan skala antar atribut.

Terdapat beberapa metode normalisasi yang sering digunakan yaitu:
1. **Min-Max Normalization** – mengubah data ke rentang 0–1

2. **Z-Score Normalization** – menggunakan mean dan standar deviasi

3. **Decimal Scaling Normalization** – membagi data dengan pangkat 10

Dengan normalisasi, proses analisis data menjadi lebih akurat dan stabil.
